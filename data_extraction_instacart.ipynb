```python
%%bash
set -e
mkdir -p data/raw
echo "Downloading Instacart from psparks dataset…"
kaggle datasets download -d psparks/instacart-market-basket-analysis -p data/raw
echo "Unzipping…"
unzip -qo data/raw/*.zip -d data/raw || true
ls -lh data/raw | sed -n '1,40p'
```

```python
import pandas as pd
import glob

csv_files = glob.glob('data/raw/*.csv')

dataframes = {}
for csv_file in csv_files:
    file_name = csv_file.split('/')[-1].replace('.csv', '')
    dataframes[file_name] = pd.read_csv(csv_file)
    print(f"Loaded {file_name}.csv with shape: {dataframes[file_name].shape}")
```

```python
for name, df in dataframes.items():
    print(f"Head of {name}.csv:")
    display(df.head())
```
